```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
import spacy
```

```python
nltk.download('stopwords')
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv('archive.zip')
```

```python
print(df.columns)
```

```python
df_cleaned = df.drop_duplicates()
```

```python
text_columns = ['Product_Name', 'Brand', 'Category', 'Skin_Type', 'Gender_Target', 'Packaging_Type', 'Main_Ingredient', 'Country_of_Origin']

# Remove special characters from these columns
for col in text_columns:
    df.loc[:, col] = df[col].astype(str).str.replace(r'\W', '', regex=True).str.strip()
```

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    doc = nlp(text)
    tokens = []
    for token in doc:
        lemma = token.lemma_.lower()
        # Include only alphabetic tokens not in stopwords
        if token.is_alpha and lemma not in stop_words:
            tokens.append(lemma)
    return ' '.join(tokens)

# Define stop_words using nltk
stop_words = set(stopwords.words('english'))

# Apply to text columns
text_columns = ['Product_Name', 'Brand', 'Category', 'Skin_Type', 'Gender_Target', 'Packaging_Type', 'Main_Ingredient', 'Country_of_Origin']
for col in text_columns:
    df[f'{col}_processed'] = df[col].apply(preprocess_text)
```
